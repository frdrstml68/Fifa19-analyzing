### Fix pictures URL definition

```python
def fix_photo_url(int_input):
    #input ID number
    output=""
    str_input=str(int_input)
    if len(str_input)<6:
        str_input="0"*(6-len(str_input))+str_input
    output="https://cdn.sofifa.com/players/"+str_input[:3]+"/"+str_input[3:]+"/19_120.png"
    return output

def fix_ClubLogo_url(str_input):
    #input ClubLogo url
    temp=str_input[37:]
    output="https://cdn.sofifa.com/teams/"+temp[:len(temp)-4]+"/60.png"
    return output

def load_dict():
    #to do
    #fill nat_dict dictionary from file: nation_flag.txt
    return True

def fix_countryFlag_url(strInput):
    #input nationality column
    output=""
    str_suffix=nat_dict[strInput]
    output="https://cdn.sofifa.com/flags/"+str_suffix+".png"
    return output
    
```

```python
#fix pics urls

data_fix=data.copy()
data_fix["Photo"]=data_fix["ID"].apply(fix_photo_url)
data_fix["Club Logo"]=data_fix["Club Logo"].apply(fix_ClubLogo_url)
data_fix["Flag"]=data_fix.Nationality.apply(fix_countryFlag_url)
```

```python
#auxiliary function
   
def players_list():
    output=data_fix.Name.tolist()
    output.sort()
    return output

#widgets objects
player_list=players_list()

player_1st = widgets.Dropdown(options=player_list,description="1st Player:")
player_2nd = widgets.Dropdown(options=player_list,description="2nd Player:")
compare_button=widgets.Button(description="Compare")
header_box=widgets.HBox([player_1st,player_2nd,compare_button])
main_box=widgets.VBox()

 #showing object
#column_data=data.index

class player_data():
    
    def __init__(self,inputData):
        """received a pd.Series as input"""
        if type(inputData)==pd.Series:
            self.data=inputData.copy()
            #print(self.data.Name)
            self.player_name=widgets.VBox()
        else:
            raise Exception("Player_data class needs a pd.Series to be initialized.")
            
    def widgt_maker(self,bln_reverse=False):
       #load imgs
        response = requests.get(self.data.Photo)
        img_pic = BytesIO(response.content).getvalue()
        response = requests.get(self.data.Flag)
        img_flag = BytesIO(response.content).getvalue()
        response = requests.get(self.data['Club Logo'])
        img_logo=BytesIO(response.content).getvalue()
        
        #player_name
        name_layout=widgets.Layout(display="flex",justify_content='center',width='250px',height='20px')
        pic_layout=widgets.Layout(display="flex",align_item='center',width='60px',height='60px')
        name_label=widgets.Label(r"\(\textbf{"+self.data.Name+"}\)",layout=name_layout)
        info1=r"\(\textbf{Age: "+str(self.data.Age)+", }\)"+r"\(\textbf{Overall: "+str(self.data.Overall)+", }\) "+r"\(\textbf{Potential: "+str(self.data.Potential)+"}\) "
        info1_label=widgets.Label(info1,layout=name_layout)
        info2=r"\(\textbf{Value: "+str(self.data.Value)+", }\)"+r"\(\textbf{Wage: "+str(self.data.Wage)+", }\) "
        info2_label=widgets.Label(info2,layout=name_layout)
        middle_box=widgets.VBox([name_label,info1_label,info2_label])
        plyr_pic=widgets.Image(value=img_pic,format='png',width=40,height=40,layout=pic_layout)
        plyr_flag=widgets.Image(value=img_flag,format='png',width=30,height=30,layout=pic_layout)
        
        if not bln_reverse:
            player_name0=widgets.HBox([plyr_flag,middle_box,plyr_pic],layout=widgets.Layout(align_item='center'))
        else:
            player_name0=widgets.HBox([plyr_pic,middle_box,plyr_flag ]) 
        
        #club part
        actual_club_Logo=widgets.Image(value=img_logo,format='png',width=60,height=60,layout=pic_layout)
        actual_club_label=widgets.Label(r"\(\textbf{"+self.data.Club+"}\)",layout=name_layout)
        info1=r"\(\textbf{Position: "+(self.data.Position)+", }\)"+r"\(\textbf{Jersey Number: "+str(self.data['Jersey Number'])+", }\)"
        actual_club_info1=widgets.Label(info1,layout=name_layout)
        info2=r"\(\textbf{Joined: "+str(self.data.Joined)+"}\) "
        actual_club_info2=widgets.Label(info2,layout=name_layout)
        actual_club_middle_box=widgets.VBox([actual_club_label,actual_club_info1])
        
        if not bln_reverse:
            player_name1=widgets.HBox([actual_club_Logo,actual_club_middle_box])
        else:
            player_name1=widgets.HBox([actual_club_middle_box,actual_club_Logo],layout=widgets.Layout(display="flex",justify_content='flex-end')) 
        
        
        #profile
      #  profile_list=['Preferred Foot', 'International Reputation', 'Weak Foot','Skill Moves', 'Work Rate', 'Body Type','Height', 'Weight', 'Real Face','Special','ID']

      #  if not bln_reverse:
          #  items_profile=[widgets.HBox([widgets.Label(r"\(\textbf{"+str(self.data[i])+ "}\)"),widgets.Label(i)]) for i in profile_list]
          #  profile=widgets.VBox(items_profile)
       # else:
        #    items_profile=[widgets.HBox([widgets.Label(r"\(\textbf{"+str(self.data[i])+ "}\)")]) for i in profile_list]
          #  profile=widgets.VBox(items_profile)    
        
        self.player_name=widgets.VBox([player_name0,player_name1],layout=widgets.Layout(display="flex",border="solid",justify_content='center'))

def profile_maker(inputData1,inputData2):
        center_layout=widgets.Layout(display="flex",justify_content='center')
        profile_list=['Preferred Foot', 'International Reputation', 'Weak Foot','Skill Moves', 'Work Rate', 'Body Type','Height', 'Weight', 'Real Face','Special','ID']
        items_profile=[widgets.HBox([widgets.Label(r"\(\textbf{"+str(inputData1[i])+ "}\)"),widgets.Label(i),widgets.Label(r"\(\textbf{"+str(inputData2[i])+ "}\)")],layout=center_layout) for i in profile_list]
        profile=widgets.VBox(items_profile)
        
        return profile

def overallRating_maker(inputData1,inputData2):
        center_layout=widgets.Layout(display="flex",justify_content='center')
        overall_list=['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
       'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM',
       'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
        items_profile=[widgets.HBox([widgets.Label(r"\(\textbf{"+str(inputData1[i])+ "}\)"),widgets.Label(i),widgets.Label(r"\(\textbf{"+str(inputData2[i])+ "}\)")],layout=center_layout) for i in overall_list]
        overall=widgets.VBox(items_profile)
        
        return overall
    
def attacking_maker(inputData1,inputData2):
        center_layout=widgets.Layout(display="flex",justify_content='center')
        attacking_list=['Crossing','Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys']
        items_profile=[widgets.HBox([widgets.Label(r"\(\textbf{"+str(inputData1[i])+ "}\)"),widgets.Label(i),widgets.Label(r"\(\textbf{"+str(inputData2[i])+ "}\)")],layout=center_layout) for i in attacking_list]
        attacking=widgets.VBox(items_profile)
        
        return attacking

def skill_maker(inputData1,inputData2):
        center_layout=widgets.Layout(display="flex",justify_content='center')
        skill_list=[ 'Dribbling','Curve', 'FKAccuracy', 'LongPassing', 'BallControl']
        items_profile=[widgets.HBox([widgets.Label(r"\(\textbf{"+str(inputData1[i])+ "}\)"),widgets.Label(i),widgets.Label(r"\(\textbf{"+str(inputData2[i])+ "}\)")],layout=center_layout) for i in skill_list]
        skill=widgets.VBox(items_profile)
        
        return skill
    
def movement_maker(inputData1,inputData2):
        center_layout=widgets.Layout(display="flex",justify_content='center')
        movement_list=['Acceleration','SprintSpeed', 'Agility', 'Reactions', 'Balance']
        items_profile=[widgets.HBox([widgets.Label(r"\(\textbf{"+str(inputData1[i])+ "}\)"),widgets.Label(i),widgets.Label(r"\(\textbf{"+str(inputData2[i])+ "}\)")],layout=center_layout) for i in movement_list]
        movement=widgets.VBox(items_profile)
        
        return movement
    
def power_maker(inputData1,inputData2):
        center_layout=widgets.Layout(display="flex",justify_content='center')
        power_list=[ 'ShotPower','Jumping', 'Stamina', 'Strength', 'LongShots']
        items_profile=[widgets.HBox([widgets.Label(r"\(\textbf{"+str(inputData1[i])+ "}\)"),widgets.Label(i),widgets.Label(r"\(\textbf{"+str(inputData2[i])+ "}\)")],layout=center_layout) for i in power_list]
        power=widgets.VBox(items_profile)
        
        return power
def mentality_maker(inputData1,inputData2):
        center_layout=widgets.Layout(display="flex",justify_content='center')
        mentality_list=[ 'Aggression','Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure']
        items_profile=[widgets.HBox([widgets.Label(r"\(\textbf{"+str(inputData1[i])+ "}\)"),widgets.Label(i),widgets.Label(r"\(\textbf{"+str(inputData2[i])+ "}\)")],layout=center_layout) for i in mentality_list]
        mentality=widgets.VBox(items_profile)
        
        return mentality
def defending_maker(inputData1,inputData2):
        center_layout=widgets.Layout(display="flex",justify_content='center')
        defending_list=['Marking', 'StandingTackle', 'SlidingTackle']
        items_profile=[widgets.HBox([widgets.Label(r"\(\textbf{"+str(inputData1[i])+ "}\)"),widgets.Label(i),widgets.Label(r"\(\textbf{"+str(inputData2[i])+ "}\)")],layout=center_layout) for i in defending_list]
        defending=widgets.VBox(items_profile)
        
        return defending
def goalkeeping_maker(inputData1,inputData2):
        center_layout=widgets.Layout(display="flex",justify_content='center')
        goalkeeping_list=['GKDiving', 'GKHandling','GKKicking', 'GKPositioning', 'GKReflexes']
        items_profile=[widgets.HBox([widgets.Label(r"\(\textbf{"+str(inputData1[i])+ "}\)"),widgets.Label(i),widgets.Label(r"\(\textbf{"+str(inputData2[i])+ "}\)")],layout=center_layout) for i in goalkeeping_list]
        goalkeeping=widgets.VBox(items_profile)
        
        return goalkeeping
        
#variable
list_player_selec=[]
 
def select_players():
    display(header_box)
    return True
  
def show_comparison(list_player):
    global main_box
    #print(list_player)
    ply1_series=data_fix.loc[data.Name==list_player[0]].squeeze()
    ply2_series=data_fix.loc[data.Name==list_player[1]].squeeze()
    center_layout=widgets.Layout(display="flex",justify_content='center')
    player_1st_data= player_data(ply1_series)
    player_1st_data.widgt_maker()
    player_2nd_data= player_data(ply2_series)
    player_2nd_data.widgt_maker(bln_reverse=True)
    profile_box=profile_maker(ply1_series,ply2_series)
    
    overall_box=overallRating_maker(ply1_series,ply2_series)
    overall_name=widgets.Label("Overall Ranking")
    
    attacking_box=attacking_maker(ply1_series,ply2_series)
    attacking_name=widgets.Label("Attacking Ranking")
    skill_box=skill_maker(ply1_series,ply2_series)
    skill_name=widgets.Label("Skill Ranking")
    movement_box=movement_maker(ply1_series,ply2_series)
    movement_name=widgets.Label("Movement Ranking")
    power_box=power_maker(ply1_series,ply2_series)
    power_name=widgets.Label("Power Ranking")
    mentality_box=mentality_maker(ply1_series,ply2_series)
    mentality_name=widgets.Label("Mentality Ranking")
    defending_box=defending_maker(ply1_series,ply2_series)
    defending_name=widgets.Label("Defending Ranking")
    goalkeeping_box=goalkeeping_maker(ply1_series,ply2_series)
    goalkeeping_name=widgets.Label("Goalkeeping Ranking")
    
   
    #display(player_1st_data.player_name)
    # player_2nd_data=widgets.VBox()
    first_row_box=widgets.Box([player_1st_data.player_name,player_2nd_data.player_name],layout=center_layout)
    main_box=widgets.VBox([first_row_box,profile_box,overall_name,overall_box, attacking_name,attacking_box,skill_name,skill_box,movement_name,movement_box,power_name,power_box,mentality_name,mentality_box,defending_name,defending_box,goalkeeping_name,goalkeeping_box],layout=center_layout)
    display(main_box)
#event handling function   

def compare_button_clicked_handled(obj):
    global list_player_selec
    global main_box
   # print("...")
    main_box.close()
    list_player_selec=[player_1st.value,player_2nd.value]
    show_comparison(list_player_selec)
   # header_box.close()

    
compare_button.on_click(compare_button_clicked_handled)
```
